## Project Description

This project trains agents to play *Celeste Classic* (the original PICO-8 version) using reinforcement learning. We compare three approaches against a random baseline and a Tool-Assisted Speedrun (TAS) upper bound:

1. **DQN** - A pure RL baseline using Double DQN with a Dueling architecture. The agent observes a 31-dimensional state (player position, velocity, dash state, and a 5×5 local tile grid) and chooses from 15 discrete actions (directional moves, jumps, and dashes).

2. **Behavioral Cloning (BC)** - Supervised imitation of TAS replay data. (state, action) pairs are exported by replaying TAS inputs through a headless PICO-8 emulator, then a classifier is trained with action-weighted cross-entropy to handle the heavy no-op class imbalance.

3. **Hybrid** - DQN warm-started from TAS data. A protected expert partition in the replay buffer preserves TAS transitions throughout online training to prevent catastrophic forgetting.


## Notebook Structure (QAC)

Each experiment section follows a **Question -> Analysis -> Conclusion** format:
- **Question:** What are we testing or measuring?
- **Analysis:** Training curves, evaluation metrics, and agent behavior.
- **Conclusion:** What did we learn, and how does it inform the next approach?
